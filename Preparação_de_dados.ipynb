```python
import pandas as pd

url = 'https://raw.githubusercontent.com/DevLucasMelloo/Transforma-o_de_dados/refs/heads/main/dados_hospedagem.json'
dados = pd.read_json(url)
dados.head()
```

# Tratando dados

```python
#Normalizando os dados
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

```python
#Lista com nome das colunas
colunas = list(dados.columns)
colunas
```

```python
#Desagrupando dados
dados = dados.explode(colunas[3:]).reset_index(drop=True)
dados.head()
```

```python
dados.info()
```

```python
import numpy as np
```

```python
#Convertendo colunas para tipo numérico inteiros
col_numericas = ['max_hospedes', 'quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
#Convertendo colunas para tipo numérico float
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
#Remoção de caracteres com replace
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
#Convertendo colunas para tipo numérico float
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
#Removendo caracteres
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
#Convertendo para float
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

# Tokenização

```python
#Transformando dados para minusculo
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados['descricao_local'][3169]
```

```python
#Removendo caracteres de texto
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
#Verificação do hifem para remoção
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
#Tokenização
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
#Remoção de caracteres
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
#Tokenização
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

# Tratando dados de Temporais

```python
url2 = 'https://raw.githubusercontent.com/DevLucasMelloo/Transforma-o_de_dados/refs/heads/main/moveis_disponiveis.json'
dt_data = pd.read_json(url2)
dt_data.head()
```

```python
dt_data.info()
```

```python
#Usando datetime para transformar em tipo data
dt_data['data'] = pd.to_datetime(dt_data['data'])
```

```python
dt_data.info()
```

```python
#Formatação de data
dt_data['data'].dt.strftime('%Y-%m')
```

# Agrupamento

```python
#Agrupando para informar quantidades de vagas por mês
subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```
